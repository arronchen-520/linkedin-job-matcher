```python
import tiktoken
import pandas as pd

encoding = tiktoken.get_encoding("cl100k_base")

def count_tokens(text):
    if not text or pd.isna(text):
        return 0
    return len(encoding.encode(str(text)))

df = pd.read_csv('../data/job_posts/filtered_posts/20260125_Arron_Machine Learning_filtered.csv')
df['real_tokens'] = df['Job Description'].apply(count_tokens)

# 看看最长的一个 JD 到底有多长
max_token = df['real_tokens'].max()
avg_token = df['real_tokens'].mean()

print(f"Max Tokens: {max_token}")
print(f"Avg Tokens: {avg_token}")

# 画个简单的直方图看看分布 (如果你在 Jupyter Notebook 里)
# df['real_tokens'].hist(bins=20)
```
